```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading in the data
team = pd.read_csv('homework_4data_2023.csv')
player = pd.read_csv('transformed_data.csv')
```

```python
# Dropping unlabeled columns
for i in player.columns[-9:]:
    player = player.drop(i, axis=1)
```

1. Use the data set provided with this homework and the Pythagorean Formula For Wins (as amended for basketball by Daryl Morey, with 14 as the exponent) to calculate how many wins the following players cost their team by virtue of their poor free throw shooting: Shaq, Dwight Howard, and Deandre Jordan.  What would have happened if they shot at a league average rate? These players’ statistics are in the first tab and their team-/league- level data are in the second tab!

```python
player['Pythagorean'] = player.apply(lambda x: x['FT']**14 /((x['FTA'] - x['FT'])**14 + x['FT']**14), axis=1)
```

```python
player
```

```python
player['Season_mod'] = player['Season'].apply(lambda x: float(x.split('-')[0]))
```

```python
years = team['Year'].unique()
res = []
for i in years:
    current_year = team.loc[team['Year'] == i]
    mean = np.mean(current_year['Tm.FT_Perc'])
    res.append({'Season_mod':i, 'FT_perc_league': mean})
```

```python
league_average = pd.DataFrame(res)
# combined = pd.merge([player, league_average], right='Season_mod')
# combined = player.merge(league_average, left_on='Season_mod', right_on='Season_mod')
```

```python
combined = player.merge(league_average, how='left', on='Season_mod')
```

```python
ax = sns.relplot(x='Season_mod', y='Pythagorean', hue='Player', data=combined)
ax._legend.remove()
ax.map_dataframe(sns.lineplot, x='Season_mod', y='FT_perc_league', color='r', label='League_average')
plt.legend(loc ="lower left")
plt.show()
plt.tight_layout();
```

What we can see is that in some years some players perform above the league free throw average such as Howard from 2005 to 2010.
